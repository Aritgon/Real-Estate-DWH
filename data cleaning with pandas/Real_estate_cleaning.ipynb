```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Soham Gon\Desktop\archive\Real_Estate_Sales_2001-2022_GL.csv")
df.head(1)
```

```python
# percentage null -
# Assessor Remarks    84.400194
#OPM remarks         98.812805

# dropping both the columns as it doesn't provide any real value to our analysis.
df.drop(columns = ['Assessor Remarks', 'OPM remarks'], inplace = True)
```

```python
df.info()
```

```python
# dropping non use code column too as it has mostly 72% of null and also doesn't provide real value.
df.drop(columns = ['Non Use Code'], inplace = True)
```

```python
# changing location to longitude and latitude.
df[['longitude', 'latitude']] = df['Location'].str.extract(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)')
```

```python
df.head(1)
```

```python
# dropping the Location column.
df.drop(columns = ['Location'], inplace= True)
```

```python
# changing the column names.
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head(1)
```

```python
df.info()
```

```python
# changing date_recorded column datatype.
df['date_recorded'] = pd.to_datetime(df['date_recorded'], errors  = 'coerce')
```

```python
df[df['property_type'] != df['residential_type']]
```

```python
# date_recorded has 0.000182% of null. its better if I impute it with mode(0)
mode_value = df['date_recorded'].mode()[0]

df['date_recorded'].fillna(mode_value, inplace = True)
```

```python
df.isnull().mean() * 100
```

```python
# address also has 0.004646 % of null. So I am also imputing it.
df['address'].fillna(df['address'].mode()[0], inplace = True)
```

```python
# checking the conflict between property_type and residential_type.
# as it has so much conflicting data. Either I have to merge or drop any of this column.

# null count - 
# property_type       34.842921
# residential_type    36.295415
```

```python
# filtering to see where exactly is the data is conflicting.
mismatched = df[df['property_type'].str.strip().str.lower() != df['residential_type'].str.strip().str.lower()]

mismatched[['property_type', 'residential_type']].value_counts()
```

```python
df['property_type'].value_counts()
```

```python
df['residential_type'].value_counts()
```

```python
matched = df[df['property_type'] == df['residential_type']]
matched[['property_type', 'residential_type']].value_counts()
```

```python
# creating a combined column using both property_type and residential_type and merging.
def match_types(property_type, residential_type):
    if property_type == 'Residential' and pd.notna(residential_type):
        return f'{property_type} - {residential_type}'
    # if null, put the property_type in empty residential_type row.
    return property_type

# create a new column and use apply() to do row-wise property_type property type + residential type data creation.
df['property_full_type'] = df.apply(lambda row: match_types(row['property_type'], row['residential_type']), axis = 1)
```

```python
df['property_full_type'].value_counts()
```

```python
df.head(5)
```

```python
# saving temporarily.
df.to_csv(r'C:\Users\Soham Gon\Desktop\archive\real_estate_cleaned().csv', index = False, encoding = 'utf-8')
```
